```python
#importing required libraries
import cv2
import matplotlib.pyplot as plt
```

```python
#taking the input from the user
face_path=input("Enter the path of the photo : ")
```

```python
#cascade classifier
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
#reading the image
img=cv2.imread(face_path)
rgb_img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#displaying the original image
print("Image before face detection")
plt.imshow(rgb_img)
```

```python
#converting to grayscale
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#getting coordinates for detected face/s
face=face_cascade.detectMultiScale(gray_img, 1.3, 5)
```

```python
#drawing a rectangle/square around the detected face/s
for(x,y,w,h) in face:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
#displaying the detetcted face/s
print("Image after face detection")
cv2.imwrite("output_scared_person.jpg", img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
